```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df = pd.read_csv('Downloads/Credit.csv').iloc[:,1:7]
df
```

### 1.Predictor: Income

```python
df.plot(kind='scatter', x = 'Limit', y = 'Income')
```

```python
plot = sns.lmplot(y='Income', x='Limit', data=df)
```

```python
r1 = sns.residplot('Limit', 'Income', data = df)
r1.set(xlabel='Lot Size', ylabel='Residuals')                                          
```

```python
lin_reg_limit = LinearRegression()
lin_reg_limit.fit(df.drop('Income', axis=1), df.iloc[:,0])
lin_reg_limit 
```

```python
lin_reg_limit.coef_
```

```python
lin_reg_limit.intercept_
```
